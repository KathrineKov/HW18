1. Output a list of tickets, where number of flights>3
To make it faster, use LIMIT 1000 or something like that.

```python
%%sql
SELECT tickets.ticket_no,
       COUNT(ticket_flights.flight_id) AS total_flights
FROM tickets
JOIN ticket_flights ON tickets.ticket_no = ticket_flights.ticket_no
GROUP BY tickets.ticket_no
HAVING COUNT(ticket_flights.flight_id) > 3
ORDER BY total_flights DESC
LIMIT 1000;
```

The output is sorted in descending order by the total number of flights, meaning tickets with the highest number of flights appear at the top. LIMIT 1000 clause helps keep the output manageable and focused on the tickets with a significant number of flights

2. Output a total number of tickets which have exactly 5 flights. Just one number, nothing else.

```python
%%sql
SELECT COUNT(*) AS total_tickets_with_five_flights
FROM (
    SELECT ticket_no
    FROM ticket_flights
    GROUP BY ticket_no
    HAVING COUNT(flight_id) = 5
) AS tickets_with_five_flights;
```

Result shows that there are 54 tickets with exactly five flights each.

This result was obtained by:
Grouping tickets by their unique ticket number
Filtering for groups (tickets) that have exactly five associated flights using the HAVING COUNT(flight_id) = 5 condition
Counting the number of such tickets to return a single summary value
